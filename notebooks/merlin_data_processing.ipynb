```python
import pandas as pd
```

```python
metadf = pd.read_csv("/Users/alaa/Documents/ucsf/data/suliman/merlin/genes_metadata.csv")
```

```python
len(metadf.columns.tolist())
```

```python
meta_df = pd.concat([metadf.iloc[:, 0], metadf.iloc[:, 41646:]], axis=1)
meta_df = meta_df.rename(columns={
    "Unnamed: 0": "sample_id",
    "PTID": "patient_id",
    "Time.point": "time",
    "group": "art_treatment_group",
    "library.pool": "library_pool",
    "pre_post": "hiv_diagnosis"
})
meta_df = meta_df.loc[~meta_df['hiv_diagnosis'].isna()].copy()
meta_df = meta_df.loc[meta_df['Code'].isin(['t0', 'tx', 'tx+1'])].copy()
print(meta_df.shape)
meta_df.head()
```

```python
meta_df.columns
```

```python
meta_df.to_csv("/Users/alaa/Documents/ucsf/data/suliman/merlin/meta/experiment1_processed_metadata_2024-12-23.csv", 
               index=False)
```
